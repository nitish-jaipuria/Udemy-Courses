```python
# simulate a linear regression

px = np.random.rand(100)
py = 3*px + 4 + np.random.rand(100)

A = ( np.mean(px*py) - np.mean(px)*np.mean(py) )/( np.mean(px**2) - np.mean(px)**2 )

B = ( np.mean(py)*np.mean(px**2) - np.mean(px)*np.mean(px*py) )/( np.mean(px**2) - np.mean(px)**2 )

r2 = 1 - (np.sum((py - (A*px+B))**2)/np.sum((py - np.mean(py))**2))
print(A,B,r2)

plt.scatter(px,py)
plt.plot(px, A*px+B, color='red')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

data = pd.read_csv("F:\\DATA SCIENCE\\Udemy\\Deep Learning Prerequisites\\machine_learning_examples-master\\linear_regression_class\\data_2d.csv",
               header = None )

data.columns = ['x1','x2', 'y']

X = data[['x1','x2']]
X['c'] = 1

X_t = np.transpose(X)

print(X.head())

y = data['y']
w = np.linalg.solve(np.dot(X_t,X), np.dot(X_t,y))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['x1'], X['x2'], y)
plt.show()

yhat = np.dot(X,w)

r2 = 1 - ( np.dot(y-yhat,y-yhat)/np.dot(y-np.mean(y),y-np.mean(y)) )

print(r2)
```
